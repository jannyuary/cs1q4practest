```python
import time
import random

def analyze_student_scores():
    passing_grade = 60 # Define passing grade
    failing_grade = 60 # Define failing grade

    student_name = input("What's your name? ")
    scores = []

    while True:
        score_input = input(f"Enter a quiz score for {student_name} (or type 'done' to finish): ")
        if score_input.lower() == 'done':
            break
        try:
            score = float(score_input)
            scores.append(score)
        except ValueError:
            print("Invalid input. Please enter a number or 'done'.")

    if not scores:
        print(f"{student_name} has no scores to analyze.")
        return

    average_score = sum(scores) / len(scores)

    print ("Loading grades...")
    time.sleep(random.randint(1, 5)) # Wait for a random duration between 1 and 5 seconds
    print(f"\n--- Student Score Analysis for {student_name} ---")
    print(f"Individual Scores: {scores}")
    print(f"Average Score: {average_score:.2f}")

    if average_score >= passing_grade:
        print(f"Result: Passed, average score is {average_score:.2f}, which is greater than or equal to the passing grade of {passing_grade} :)!")
    else:
        print(f"Result: Failed, average score is {average_score:.2f}, which is less than the failing grade of {failing_grade} :(")

# Call the function to run the program
analyze_student_scores()
```
